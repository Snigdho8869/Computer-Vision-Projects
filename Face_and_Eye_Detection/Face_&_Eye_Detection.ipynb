#Importing the Necessary Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

#Displaying the Image

```python
image_path = '/content/chris-evans-gettyimages-1138769185.jpg'

plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.axis('off')
plt.show()
```

# Initializing Face and Eye Detection Classifiers

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

eye_detector = cv2.CascadeClassifier('haarcascade_eye.xml')
```

#Visualizing Detected Faces in an Image

```python
def detect_face(image_path, min_face_size=(60, 60)):
    image = cv2.imread(image_path)
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


    faces = face_detector.detectMultiScale(grayscale_image, 1.1, 4)

    for (x, y, w, h) in faces:
        if w >= min_face_size[0] and h >= min_face_size[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 10)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Image with Detected Faces')
    plt.axis('off')
    plt.show()
```

```python
detect_face(image_path)
```

#Visualizing Detected Eyes in an Image

```python
def detect_eye(image_path, min_eye_size=(40, 40)):
    # Load the image
    image = cv2.imread(image_path)

    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


    eyes = eye_detector.detectMultiScale(grayscale_image, 1.1, 4)

    for (x, y, w, h) in eyes:
        if w >= min_eye_size[0] and h >= min_eye_size[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 6)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Image with Detected Eyes')
    plt.axis('off')
    plt.show()
```

```python
detect_eye(image_path)
```

# Visualizing Face and Eye Detection in an Image

```python
def detect_face_and_eyes(image_path, min_face_size=(60, 60), min_eye_size=(40, 40)):

    image = cv2.imread(image_path)

    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faces = face_detector.detectMultiScale(grayscale_image, 1.1, 4)

    for (x, y, w, h) in faces:
        if w > min_face_size[0] and h > min_face_size[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 10)

            roi = grayscale_image[y:y + h, x:x + w]
            eyes = eye_detector.detectMultiScale(roi, 1.1, 4)

            for (ex, ey, ew, eh) in eyes:
                if ew > min_eye_size[0] and eh > min_eye_size[1]:
                    cv2.rectangle(image, (x + ex, y + ey), (x + ex + ew, y + ey + eh), (0, 0, 255), 6)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Image with Detected Faces and Eyes')
    plt.axis('off')
    plt.show()
```

```python
detect_face_and_eyes(image_path)
```
